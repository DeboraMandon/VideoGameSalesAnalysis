```python
from tqdm import tqdm # librairie permettant d'indiquer le pourcentage de la boucle for
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install()) # Ouvrir une page Google Chrome
# créer une attente implicite
driver.implicitly_wait(10)

gaming_wiki=gaming2

driver.get('https://fr.wikipedia.org/wiki/Wikip%C3%A9dia:Accueil_principal') # Ouvrir l'adresse du site web

#Lancer une recherche en utilisant un fichier excel ou csv (en ittérant sur une colonne)

for i, row in tqdm(gaming_wiki.iterrows(), total = len(gaming_wiki)):
        
    driver.find_element(By.ID, "searchform") # acceder à la barre de recherche de Wikipedia
    bouton_recherche=driver.find_element(By.ID, "searchform")
    sleep(1) 
    bouton_recherche.click()
    sleep(1) 
    driver.find_element(By.CLASS_NAME, "cdx-text-input__input")
    barre_rech_text=driver.find_element(By.CLASS_NAME, "cdx-text-input__input")
    barre_rech_text.click()
    barre_rech_text.send_keys(row['Name']) # Entrer un texte dans la barre de recherche
    barre_rech_text.send_keys(Keys.ENTER) # Cliquer sur Entrer
    
    try:
        info_box = driver.find_element(By.XPATH, "//*[@id='mw-content-text']/div[1]/div[2]")
        gaming_wiki.loc[i, 'description_wiki'] = info_box.text
    except:
        gaming_wiki.loc[i, 'description_wiki'] = 'no_data'
```

```python
driver.close()
```
