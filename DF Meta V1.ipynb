```python
import pandas as pd
```

```python
df1 = pd.read_csv('all_games.csv')
df2 = pd.read_csv('vgsales.csv')
```

```python
display(df1.head())
display(df2.head())
```

```python
#On renomme les colonnes du df1 contenant les scores de metacritic

df1 = df1.rename(columns={"name": "Name", "release_date": "Year", 'platform':'Platform'})
df1
```

```python
display(df1['Platform'].unique())

display(df2['Platform'].unique())
```

```python
Platform_name = {' Nintendo 64':'N64',' PlayStation':'PS',' PlayStation 3':'PS3',' Dreamcast':'DC',' Xbox 360':'X360',
               ' Wii':'Wii',' Xbox One':' XOne',' PC':'PC',' PlayStation 2':'PS2',' PlayStation 4':'PS4',
               ' GameCube':'GC',' Xbox':'XB',' Wii U':'WiiU',' Game Boy Advance':'GBA',' 3DS':'3DS',' DS':'DS',
               ' PlayStation Vita':'PSV',' PSP':'PSP',' Stadia':'SAT'}
```

```python
df1['Platform'].replace(Platform_name,inplace = True)
df1
```

```python
#convertion en format date 
df1['Year'] = pd.to_datetime(df1['Year'])
```

```python
#extraction de l'année avec la fonction year
df1['year'] = df1['Year'].dt.year
```

```python
df2.dropna(inplace = True)
```

```python
df2['year'] = df2['Year'].astype(int)
df2
```

```python
#suppression de la colonne Year mauvaise
df2 = df2.drop(['Year'], axis=1)
```

```python
df2
```

```python
df1 = df1.drop(['Year'], axis=1)
```

```python
df1
```

```python
#merge de metagames et vgsales
df_merge = df1.merge(df2,on=['Name','year','Platform'],how='inner')
```

```python
df_merge
```

```python
df_merge.info()
```

```python
df_merge.to_csv('df_metagames.csv')
```

```python
plt.figure(figsize=(15,5))
valeurs=top_10_user[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum()
plt.title('Répartition de vente des 10 meilleurs avis utilisateurs par Région')
plt.pie(
    x=valeurs.values,
    labels=valeurs.index,
    autopct=lambda x: f'{str(round(x, 2))}%',
);

plt.figure(figsize=(15,5))
valeurs=top_10_meta[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum()
plt.title('Répartition de vente des 10 meilleurs avis meta par Région')
plt.pie(
    x=valeurs.values,
    labels=valeurs.index,
    autopct=lambda x: f'{str(round(x, 2))}%',
);
```

```python
top_user = df_merge.sort_values(['user_review'], ascending=False)
display(top_user[['Name','user_review','year','Genre','Publisher','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']])

top_meta = df_merge.sort_values(['meta_score'], ascending=False)
display(top_meta[['Name','meta_score','year','Genre','Publisher','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']])

```

```python
NA_Genre=top_user.groupby("Genre")[["NA_Sales"]].sum().sort_values(by=['NA_Sales'],ascending=[False]).reset_index()
EU_Genre=top_user.groupby("Genre")[["EU_Sales"]].sum().sort_values(by=['EU_Sales'],ascending=[False]).reset_index()
JP_Genre=top_user.groupby("Genre")[["JP_Sales"]].sum().sort_values(by=['JP_Sales'],ascending=[False]).reset_index()
Other_Genre=top_user.groupby("Genre")[["Other_Sales"]].sum().sort_values(by=['Other_Sales'],ascending=[False]).reset_index()
Global_Genre=top_user.groupby("Genre")[["Global_Sales"]].sum().sort_values(by=['Global_Sales'],ascending=[False]).reset_index()


fig, (ax1,ax2,ax3,ax4,ax5) = plt.subplots(5,1, figsize=(8, 16), sharey=True)
fig.subplots_adjust(hspace=0.5)

sns.barplot(x='Genre', y='NA_Sales',
            palette="viridis",
            data=NA_Genre,
            ax=ax1)
ax1.set_xticklabels(ax1.get_xticklabels(),
                    rotation=25, ha='right')

sns.barplot(x='Genre', y='EU_Sales',
            palette="viridis",             
            data=EU_Genre,
            ax=ax2)
ax2.set_xticklabels(ax2.get_xticklabels(),
                    rotation=25, ha='right')

sns.barplot(x='Genre', y='JP_Sales',
            palette="viridis", 
            data=JP_Genre,
            ax=ax3)
ax3.set_xticklabels(ax3.get_xticklabels(),
                    rotation=25, ha='right')

sns.barplot(x='Genre', y='Other_Sales',
            palette="viridis",
            data=Other_Genre,
            ax=ax4)
ax4.set_xticklabels(ax4.get_xticklabels(),
                    rotation=25, ha='right')

sns.barplot(x='Genre', y='Global_Sales',
            palette="viridis",
            data=Global_Genre,
            ax=ax5);
ax5.set_xticklabels(ax5.get_xticklabels(),
                    rotation=25, ha='right');
```

```python
NA_Genre=top_meta.groupby("Genre")[["NA_Sales"]].sum().sort_values(by=['NA_Sales'],ascending=[False]).reset_index()
EU_Genre=top_meta.groupby("Genre")[["EU_Sales"]].sum().sort_values(by=['EU_Sales'],ascending=[False]).reset_index()
JP_Genre=top_meta.groupby("Genre")[["JP_Sales"]].sum().sort_values(by=['JP_Sales'],ascending=[False]).reset_index()
Other_Genre=top_meta.groupby("Genre")[["Other_Sales"]].sum().sort_values(by=['Other_Sales'],ascending=[False]).reset_index()
Global_Genre=top_meta.groupby("Genre")[["Global_Sales"]].sum().sort_values(by=['Global_Sales'],ascending=[False]).reset_index()


ig, (ax1,ax2,ax3,ax4,ax5) = plt.subplots(5,1, figsize=(8, 16), sharey=True)
fig.subplots_adjust(hspace=0.5)

sns.barplot(x='Genre', y='NA_Sales',
            palette="viridis",
            data=NA_Genre,
            ax=ax1)
ax1.set_xticklabels(ax1.get_xticklabels(),
                    rotation=25, ha='right')

sns.barplot(x='Genre', y='EU_Sales',
            palette="viridis",             
            data=EU_Genre,
            ax=ax2)
ax2.set_xticklabels(ax2.get_xticklabels(),
                    rotation=25, ha='right')

sns.barplot(x='Genre', y='JP_Sales',
            palette="viridis", 
            data=JP_Genre,
            ax=ax3)
ax3.set_xticklabels(ax3.get_xticklabels(),
                    rotation=25, ha='right')

sns.barplot(x='Genre', y='Other_Sales',
            palette="viridis",
            data=Other_Genre,
            ax=ax4)
ax4.set_xticklabels(ax4.get_xticklabels(),
                    rotation=25, ha='right')

sns.barplot(x='Genre', y='Global_Sales',
            palette="viridis",
            data=Global_Genre,
            ax=ax5);
ax5.set_xticklabels(ax5.get_xticklabels(),
                    rotation=25, ha='right');
```
