```python
# IMPORT DES MODULES ET DU DATASET 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

path="https://raw.githubusercontent.com/DeboraMandon/video_game_sales_analysis/main/vgsales.csv"
df=pd.read_csv(path, index_col=0)
df.head()
```

```python
# 1/ répartition des jeux par catégorie
ax = sns.countplot(x='Genre',data=df,order=df['Genre'].value_counts().index)
plt.xticks(rotation=45);
plt.ylabel('Nb de jeux vendus')
plt.title('Distribution du nombre de jeu par catégorie');

total = len(df)
# code honteusement pompé de l'internet !
# Mais ça a le mérite de fonctionner et de rajouter les % au dessus de chaque barre
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%\n'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='center')
```

```python
# 1 - BIS / répartition des ventes par catégorie
df_SC=df.groupby(['Genre']).agg({'Global_Sales':sum}).sort_values(by='Global_Sales',ascending=False)

ax2 = sns.barplot(x=df_SC.index,y='Global_Sales',data=df_SC)
plt.xticks(rotation=45);
plt.ylabel('Total des Ventes en M d"€')
plt.title('Distribution des ventes en € par catégorie');

total2 = df_SC.Global_Sales.sum()
# code honteusement pompé de l'internet !
# Mais ça a le mérite de fonctionner et de rajouter les % au dessus de chaque barre
for t in ax2.patches:
    percentage = f'{100 * t.get_height() / total2:.1f}%\n'
    x = t.get_x() + t.get_width() / 2
    y = t.get_height()
    ax2.annotate(percentage, (x, y), ha='center', va='center')

# On observe que la répartition reste globalement la même, il n'y a pas d'écart flagrant
```

```python
# 2/ Répartition des ventes  par an (et par plateforme ?)

# changement de 'YEAR' au format date
df['Year']=pd.to_datetime(df['Year'], format='%Y')
df_year=df.groupby(['Year']).agg({'NA_Sales':sum, 'EU_Sales':sum, 'JP_Sales':sum, 'Other_Sales':sum,'Global_Sales':sum})

# on teste une boucle for sur des graphs :D
# liste des noms des colonnes : df_year.columns
plt.figure(figsize=(10,5))
for colonne in df_year.columns :
    plt.plot_date(x=df_year.index,y=df_year[colonne],xdate=True,linestyle='-',label=colonne);
plt.grid()
plt.title('Répartition des ventes par année et par Région')
plt.xlabel('Année de sortie')
plt.ylabel('Ventes en M €')
plt.legend();

#ça marche :D aahahah
```

```python
# 2 BIS / Nb de jeux sortis par an (et par plateforme ?)

# changement de 'YEAR' au format date
#df['Year']=pd.to_datetime(df['Year'], format='%Y')
df2_year=df['Year'].value_counts().sort_index()

plt.figure(figsize=(10,5))
plt.plot_date(x=df2_year.index,y=df2_year.values,xdate=True,linestyle='-.');
plt.grid()
plt.title('Nombre de jeux sortis par an')
plt.xlabel('Année de sortie')
plt.ylabel('Nombre de jeux');

# Graph bizarre et incohérent après 2008 ??? 
# Le nombre de jeux sortis par an ne peut pas chuter comme cela
# on peut se poser 2 questions : 
    # 1/ les données que nous avons dans ce dataset sont elles fiables ?
    # 2/ d'où pourraient provenir ces données ?
```

```python
# 4/ Répartition des Sales par catégorie / Publisher / plateforme
plt.figure(figsize=(15,5))
ax = sns.countplot(x='Platform',data=df,order=df['Platform'].value_counts().index)
plt.xticks(rotation=45);
plt.ylabel('Nb de jeux vendus')
plt.title('Répartition du nombre de jeux sortis par Plateforme');

total = len(df)
# code honteusement pompé de l'internet !
# Mais ça a le mérite de fonctionner et de rajouter les % au dessus de chaque barre
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%\n'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='center')
```

```python
# 5/ Répartition des Sales par Région et par an ?
plt.figure(figsize=(15,5))
valeurs=df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum()
plt.title('Répartition des Sales par Région')
plt.pie(x=valeurs.values,labels=valeurs.index,autopct=lambda x : str(round(x,2))+'%');
```
