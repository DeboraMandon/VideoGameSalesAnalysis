```python
# IMPORT DES MODULES ET DU DATASET 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

path="https://raw.githubusercontent.com/DeboraMandon/video_game_sales_analysis/main/vgsales.csv"
df=pd.read_csv(path, index_col=0)
df.head()
```

```python
# AFFICHAGE DU TYPE DES COLONNES DU DATASET

df.info()
```

```python
# DESCRIBE SUR LES VARIABLES CONTINUES

df.describe()
```

```python
df.head(20)
#idée graphiques:
# 1/ répartition des jeux par catégorie
# 2/ Nb de jeux sortis par an (et par plateforme ?)
# 3/ Nb de jeux sortits par an et par Publisher
# 4/ Répartition des Sales par catégorie / Publisher / plateforme
# 5/ Répartition des Sales par Région
```

```python
# 1/ répartition des jeux par catégorie
ax = sns.countplot(x='Genre',data=df,order=df['Genre'].value_counts().index)
plt.xticks(rotation=45);
plt.ylabel('Nb de jeux vendus')
plt.title('Distribution du nombre de jeu par catégorie');

total = len(df)
# code honteusement pompé de l'internet !
# Mais ça a le mérite de fonctionner et de rajouter les % au dessus de chaque barre
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%\n'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='center')
```

```python
# 2/ Nb de jeux sortis par an (et par plateforme ?)
plt.plot_date(x=df.Year,y=df.Global_Sales,xdate=True);
```

```python
# 4/ Répartition des Sales par catégorie / Publisher / plateforme
plt.figure(figsize=(15,5))
ax = sns.countplot(x='Platform',data=df,order=df['Platform'].value_counts().index)
plt.xticks(rotation=45);
plt.ylabel('Nb de jeux vendus')
plt.title('Répartition du nombre de jeux sortis par Plateforme');

total = len(df)
# code honteusement pompé de l'internet !
# Mais ça a le mérite de fonctionner et de rajouter les % au dessus de chaque barre
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%\n'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='center')
```
